9. Write a paragraph to highlight the differences between the fill foward/backward versus linear interpolation of the missing values. 



### Answers
Both fill and linear interpolation helps to fill the missing values, but each does in a different way. Fill method fills the missing gap using the last known value forward (or backward) assuming nothing has changed. Linear interpolation on the other hand does'nt just copies and pastes the last known value, it reads between the lines, drawing a straight-line bridge between the last known value before the gap and the first one after, reulting a more natural than the fill method.

10. Time series interpolation works by calling interpolate with the option method='time'.
    - Explain how is it different than the linear interpolation? 
    - when do linear and time interpolation give the same results? `

### Answers

Time Interpolation - Explicitly uses the actual time intervals between data points in a datetime index. Interpolates values based on the proportional time elapsed between existing points.
Linear Interpolation - Treats the index as numerical positions (e.g., integers or converted datetime values like nanoseconds since the epoch). For datetime indices, this is mathematically equivalent to time interpolation because the numerical representation inherently reflects time intervals.


linear and time interpolation give the same results the time intervals between data points are uniform (e.g., hourly data with no gaps).

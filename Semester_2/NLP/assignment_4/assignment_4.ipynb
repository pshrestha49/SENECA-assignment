```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

# Load corpus from file
file_path = "corpus5.txt"
with open(file_path, 'r') as f:
    documents = [line.strip() for line in f if line.strip()]


# Frequency Representation
freq_vectorizer = CountVectorizer(lowercase=True, stop_words='english')
freq_matrix = freq_vectorizer.fit_transform(documents)
freq_df = pd.DataFrame(freq_matrix.toarray(), columns=freq_vectorizer.get_feature_names_out())
print("Frequency Representation")
print(freq_df)


# TF-IDF Representation
tfidf_vectorizer = TfidfVectorizer(lowercase=True, stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(documents)
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())
print("TF-IDF Representation")
print(tfidf_df)
```

```python
from gensim.models import KeyedVectors
 
# Path to the pre-trained GloVe word vectors file
glove_file = 'glove.6B.100d.txt'
 
# Load GloVe word vectors into Gensim KeyedVectors format
word_vectors = KeyedVectors.load_word2vec_format(glove_file, binary=False)
```

```python
word_vectors.vectors
```

```python
import re

# Load corpus
with open('corpus5.txt', 'r', encoding='utf8') as f:
    corpus = f.read()

# Tokenize: lowercase, remove punctuation, split words
def tokenize(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)  
    return text.split()

tokens = tokenize(corpus)

documents = corpus.strip().split('\n\n')
first_doc_tokens = tokenize(documents[0])
```

```python
for word in first_doc_tokens:
    if word in word_vectors:
        print(f"{word}: {word_vectors[word]}")
    else:
        print(f"{word}: <not in vocabulary>")
```

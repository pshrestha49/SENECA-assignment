9. Write a paragraph to highlight the differences between the fill foward/backward versus linear interpolation of the missing values.



Fill forward/backward method will cover all values, and therefore remove all nulls. The edge cases (ie. Beginning rows if the first null value is index 0, and the end rows if the last null value is index -1) are handled through backfill (which will fill with the NEXT known value, removing any nulls from the beginning), and forwardfill (which will fill with the LAST known value, removing any nulls from the end.

Linear interpolation, by default, will repeat the last known value for the end fo the dataframe, and will return NaN for the first NaN values at the top.

Linear Interpolation simply means to estimate a missing value by connecting dots in a straight line in increasing order. In short, It estimates the unknown value in the same increasing order from previous values. The default method used by Interpolation is Linear. So while applying it, we need not specify it. If there are no values to begin with (ie. if the first value is NaN), then there is to starting point to connect our dots from, causing a NaN value.

10. Time series interpolation works by calling interpolate with the option method='time'.
    - Explain how is it different than the linear interpolation?
    - when do linear and time interpolation give the same results?

Linear interpolation will ignore the index, and assume that every datapoint is evenly spaced out.

Time-Series interpolation will not ignore the index, and will use it to determine how far apart the points are (ie. time series data would have a date as the index - it would assume if the index is separated by days, that we need to account for number of days between two non-missing values.)

They will therefore, give the same results if we have a complete index (ie. our time series data, has rows for every single day in a time period). Linear interpolation can ignore the index since it assumes all datapoints are equally spread out, and timeseries will know that all the datapoints actually ARE evenly spread out.

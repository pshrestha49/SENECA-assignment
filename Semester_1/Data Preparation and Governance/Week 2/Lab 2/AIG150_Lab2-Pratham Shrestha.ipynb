16. Pivot the dataframe to hold information in the following format:
class 	                                  First 	Second  	  Third
sex 			
female  	                              0.968085  	0.921053 	0.500000
male 	                                  0.368852  	0.157407 	0.135447

```python
titanic_df_survived = titanic_df.pivot_table(index='sex', columns='class',values='survived')
titanic_df_survived
```

17. Pivot the table further to hold information in the following format:
     	    class 	First 	Second 	Third
sex 	age 			
female 	(0, 18] 	0.909091 	1.000000 	0.511628
		(18, 80] 	0.972973 	0.900000 	0.423729
male 	(0, 18] 	0.800000 	0.600000 	0.215686
		(18, 80] 	0.375000 	0.071429 	0.133663

```python
titianic_pivot_table = titanic_df.pivot_table(
    index=["sex", pd.cut(titanic_df["age"], bins=[0, 18, 80], labels=["(0, 18]", "(18, 80]"])],
    columns="class",
    values="survived",
)
# by default the agg function 
titianic_pivot_table
```

18.Pivot the table further to hold information in the following format:
fare 	survived
class 	First 	Second 	Third 	First 	Second 	Third
sex 						
female 	106.125798 	21.970121 	16.118810 	91.0 	70.0 	72.0
male 	67.226127 	19.741782 	12.661633 	45.0 	17.0 	47.0

```python
titanic_fare_pivot_df = titanic_df.pivot_table(
    index=["sex"],
    columns="class",
    values=["fare", "survived"],
    aggfunc={"fare":"mean", "survived":"sum"}
)
titanic_fare_pivot_df
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### 1. Importing Data and Matplotlib

```python
df = pd.read_csv("company_sales_data.csv")
df
```

### 2. Creating Line Plots

```python
x = df['month_number']
y = df['total_profit']

plt.plot(x, y, marker='o', color='g')
plt.xlabel("Month Number") 
plt.ylabel("Total Profit") 
plt.title("Monthly Total Profit") 
plt.grid(True)
plt.style.use("Solarize_Light2")

plt.show()
```

### 3. Multiple Line Plots

```python
y1 = df['facewash']
y2 = df['bathingsoap']

plt.plot(x, y1,  color='crimson', label='Face wash')
plt.plot(x, y2,  label='Bathing soap')
plt.legend()
plt.style.use("Solarize_Light2")
plt.show()
```

### 4. Bar Plots

```python
plt.bar(x, y, color='lightgreen')
plt.xlabel("Month"), plt.ylabel("Total Profit") 
plt.title("Profit per Month") 

plt.show()
```

### 5. Scatter Plots

```python
x = df['facecream']
y = df['total_profit']

plt.scatter(x, y, color='blue', marker='o')
plt.xlabel("Face Cream Units")
plt.ylabel("Total Profit")
plt.grid(True)
plt.style.use("Solarize_Light2")
plt.show()
```

### 6. Pie Chart

```python
labels = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
values = [df[clmn].sum() for clmn in labels]

plt.figure(figsize=(12, 12))
plt.pie(values, labels=labels, autopct='%1.1f%%')
plt.title("Piechart for Percentage of Total Sales by Product")
plt.style.use("Solarize_Light2")
plt.legend()
plt.show()
```

### 7. Histogram

```python
data = df['total_profit']

plt.figure(figsize=(10, 10))
plt.hist(data, bins=6, color='cyan', edgecolor = 'black')
plt.xlabel("Monthly Total Profit")
plt.ylabel("Frequency")
plt.title("Total Profit Distribution per month")
plt.style.use("Solarize_Light2")
plt.show()
```

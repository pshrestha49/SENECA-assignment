```python
#5. For male observations, replace the missing values under the age column with the mean age of all male observations.
```

```python
#6. Write the code to show the rows of the stroke dataframe where there are outliers for the avg_glucose_level variable.
#   Outliers should be considered based on the mean and standard deviation method.Any value greater than the mean +2*SD or
#  lower than mean-2*SD is considered as outlier.
```

```python
mean_glucose_level = stroke['avg_glucose_level'].mean()
standard_glucose_level = stroke['avg_glucose_level'].std()

for i_index, i_row in stroke.iterrows():
    if (i_row['avg_glucose_level']> (mean_glucose_level + 2*standard_glucose_level)) or (i_row['avg_glucose_level']< (mean_glucose_level - 2*standard_glucose_level)):
        print(f'outlier row --> {i_row}')
        print(i_index)
```

```python
#8.Write a paragraph to highlight the differences between the fill foward/backward versus linear interpolation of the missing values.
```

Forward fill (ffill) and backward fill (bfill) propagate the last known or next known value, respectively, to fill missing values. These methods work well for time-series data where missing values are assumed to follow the previous trend. However, they do not estimate intermediate values based on a trend. In contrast, linear interpolation calculates missing values by assuming a straight-line relationship between known data points. It provides a more accurate estimate when data follows a continuous trend, unlike ffill or bfill, which may introduce bias by repeating values. Linear interpolation is ideal for numerical datasets with gradual changes, while forward and backward fill are better suited for categorical or sequential data where trends remain consistent over time.
